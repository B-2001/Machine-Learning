# **Unsupervised Learning on Country Data**

* Dataset link: [https://www.kaggle.com/datasets/rohan0301/unsupervised-learning-on-country-data/data?select=Country-data.csv](https://www.kaggle.com/datasets/rohan0301/unsupervised-learning-on-country-data/data?select=Country-data.csv)

* in this project, we are going to classify countries into 3 categories - developed country, developing country, poor (Least Developed) based on economic and health factors from the country dataset.

* Dataset Description:

    | **Column Name** | **Description** |
    |-----------------|-----------------|
    | `country`       | Name of the country |
    | `child_mort`    | Death of children under 5 years of age per 1000 live births |
    | `exports`       | Exports of goods and services per capita, given as % of the GDP per capita |
    | `health`        | Total health spending per capita, given as % of GDP per capita |
    | `imports`       | Imports of goods and services per capita, given as % of the GDP per capita |
    | `income`        | Net income per person |
    | `inflation`     | The measurement of the annual growth rate of the Total GDP |
    | `life_expec`    | The average number of years a newborn child would live if current mortality patterns remain the same |
    | `total_fer`     | The number of children that would be born to each woman if the current age-fertility rates remain the same |
    | `gdpp`          | The GDP per capita, calculated as the Total GDP divided by the total population |

### Import neccessary libraries

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as exp
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
```

# # EDA

```python
df.corr()
```

```python
fig = plt.figure(figsize=(12,10))
ax = sns.heatmap(df.corr(),annot=True,cmap = 'viridis')
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler
scalar = MinMaxScaler()
```

```python
scaled_data = scalar.fit_transform(df.drop('country',axis = 1))
```

```python
scaled_df = pd.DataFrame(data = scaled_data,columns=df.columns[1:])
scaled_df['country'] = df['country']
scaled_df
```

```python
exp.histogram(data_frame=df,x = 'gdpp',nbins=167,opacity=0.75,barmode='overlay')
```

```python
exp.scatter(data_frame=df,x = 'child_mort',y = 'health',color='country')
```

```python
data = scaled_df.drop('country',axis = 1)
```

```python
# Calculate sum of squared distances
ssd = []
K = range(1,10)
for k in K:
    km = KMeans(n_clusters=k)
    km = km.fit(data)
    ssd.append(km.inertia_) 
```

```python
plt.figure(figsize=(10,6))
plt.plot(K, ssd, 'bx-')
plt.xlabel('k')
plt.ylabel('ssd')
plt.title('Elbow Method For Optimal k')
plt.show()
```

# # Kmeans Clustering implemendation

```python
kmean = KMeans(n_clusters=3)
kmean.fit(data)
```

```python
pred = kmean.labels_
print(pred)
```

```python
exp.scatter(data_frame= df,x = 'gdpp',y = 'income',color=kmean.labels_)
```

# # PCA implemendation

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
pca_model = pca.fit_transform(data)
data_transform = pd.DataFrame(data = pca_model, columns = ['PCA1', 'PCA2'])
data_transform['Cluster'] = pred
```

```python
plt.figure(figsize=(8,8))
g = sns.scatterplot(data=data_transform, x='PCA1', y='PCA2', palette=sns.color_palette()[:3], hue='Cluster')
title = plt.title('Countries Clusters with PCA')
```
